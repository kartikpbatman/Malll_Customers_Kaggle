```python
import pandas as pd
import seaborn as sns
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

# ANALYZE THE DATA

```python
data_dt=pd.read_csv("Mall_Customers.csv")
data_dt.head(10)
```

```python
data_dt.info()
```

# FIND A RELATION BETWEEN SPENDING SCORE AND OTHER COLS

```python
data_dt.corr(numeric_only=True)['Spending Score (1-100)'].sort_values
```

## SEARCH FOR NULL VALUES

```python
null_Values=data_dt.isnull().sum()
null_Values=null_Values[null_Values>0]
```

```python
plt.figure(figsize=(8,7))
null_Values.plot(kind='bar') ## no missing values so no graph
```

## only one impacted feature

```python
graph_A=data_dt[['Annual Income (k$)']]
```

# plot the data and determine clusters

```python
plt.figure(figsize=(10,5))
for i, feature in enumerate(graph_A):
    sns.scatterplot(data=data_dt,x=feature,y='Spending Score (1-100)',alpha=0.7)
    plt.title(f"{feature} vs Spending Score (1-100)")
```

# Loop For values of K

```python
Impacts_A=data_dt[['Annual Income (k$)','Spending Score (1-100)']]
wcss=[]
for i in range (1,10):
    kmeans=KMeans(n_clusters=i,init='k-means++',random_state=35)
    kmeans.fit(Impact_A)
    wcss.append(kmeans.inertia_)
```

## Determine the best K

```python
plt.plot(range(1,10),wcss,marker='p')
plt.title('best K')
plt.xlabel('clusters')
plt.ylabel('wcss')
plt.show()
```

```python
k=5
kmeans=KMeans(n_clusters=k,init='k-means++',random_state=35)
Kmeans=kmeans.fit_predict(Impacts_A)
```

## Attach the clusters to csv

```python
data_dt['clusters']=Kmeans
data_dt.head(10) 
```

```python
centers=kmeans.cluster_centers_
centers_dt=pd.DataFrame(centers,columns=['Annual Income (k$)','Spending Score (1-100)'])
print(centers_dt)
```

## Visualize the clusters

```python
color=['green','yellow','blue','purple','red']
for i in range(k):
    plt.scatter(Impacts_A.iloc[Kmeans==i,0],Impacts_A.iloc[Kmeans==i,1],s=80,c=color[i],label=f'cluster{i}')

plt.scatter(centers[:,0],centers[:,1],s=250,c='black',marker='.',label='Centroids')
plt.title('Mall customers')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.show()
```

```python
clustered_data = Impacts_A.copy()
clustered_data['cluster'] = Kmeans
```

```python
print(clustered_data.groupby('cluster').mean())
print(clustered_data["cluster"].value_counts())
```
